# Data Cleaning:

```python
df.isnull().sum()
```

```python
# calculate the interquartile range (IQR)
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

# define the outlier limits
lower_outlier_limit = q1 - 1.5 * iqr
upper_outlier_limit = q3 + 1.5 * iqr

# identify outliers
outliers = df[(df < lower_outlier_limit) | (df > upper_outlier_limit)]
```

# Data visualisation and Analysis:

```python
plt.boxplot(df['Annual_Premium'])
plt.title('Box Plot of Annual Premium')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
df.info()
```

```python
# Histogram to understand the distribution of numerical variables
plt.figure(figsize=(12,6))
sns.distplot(df['Region_Code'], bins=30)
plt.title('Distribution of Numerical Column')
plt.show()
```

```python
# Histogram Visualization of all the numeric columns
df.hist(column=['Age','Region_Code',
                      'Vintage','Policy_Sales_Channel',
                      'Annual_Premium', 'Previously_Insured'], 
              color='steelblue', # many options, some to consider are
                            # ['aqua', 'red', 'gold', 'royalblue', 
                            #'darkorange', 'green', 'purple', 
                            #'steelblue', 'yellow', 'lime', 'magenta']
              #edgecolor='black', # color of the edge line for the bars
              figsize=(10,15), #(x axis and y axis in inches)
              bins=10, # number of bins to divide the data into
              rwidth=0.9, # width between each bin
             )
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Gender', y='Region_Code')
plt.title('Boxplot of Numerical Column vs Categorical Column')
plt.show()
```

```python
# Pairplot to visualize the relationship between variables
sns.pairplot(df)
plt.title('Pairplot')
plt.show()
```

```python
# Heatmap to visualize the correlation between numerical variables
correlation_matrix = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.info()
```

```python
# Scatterplot for visualizing the relationship between id and Annual Premium
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='id', y='Annual_Premium')
plt.title('Scatterplot of id vs Annual Premium')
plt.show()
```

```python
#Horizontal bar graph to visualize customers Vehicle's Age
ax = df.Vehicle_Age.value_counts().plot(kind='barh', figsize=(10,2), grid=False)
ax.bar_label(ax.containers[0], label_type='center', color='white')
plt.title('Vehicle Age of the Customer')
plt.xlabel('Vehicle Age')
plt.show()
```

```python
#Pie chart to visualize the percentage of customers who are willing to purchase insurance
ax = df.Response.value_counts().plot(kind='pie', figsize=(10,3), 
                                        grid=False, ylabel='',
                                        labels = ['No', 'Yes'],
                                        shadow=True, explode=[0.1, 0.05],
                                        legend=True, labeldistance=None, 
                                        autopct='%1.2f%%')
ax.legend(bbox_to_anchor=(1, 1),loc='upper left')
plt.title('Customers willing to purchase Insurance')
plt.show()
```

```python
# Boxplot for visualizing the distribution of target variable values
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Gender', y='Policy_Sales_Channel')
plt.title('Boxplot of Policy Sales Channel vs Gender')
plt.show()
```

```python
# violin plot to visualize the distribution of the data with response as a target variable
plt.figure(figsize=(15,10))
for i,col in enumerate(df.columns,1):
    if col=='Response': continue
    plt.subplot(3,4,i)
    plt.title(f"Distribution of {col} Data")
    sns.violinplot(data=df, x='Response', y=df[col])
    plt.tight_layout()
    plt.plot()
```

```python
#Distribution of age within the dataset using a histogram or density plot
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='Age', element="step", stat="density")
plt.title('Histogram of Age')
plt.show()
```
